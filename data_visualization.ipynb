```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from rdkit.Chem import Draw
import rdkit
```

Setting the Dataset

```python
DATASET_PATH=r"C:/Tg_Bayesian_Optimization/datasets/TgSS_enriched_cleaned.csv"
df=pd.read_csv(DATASET_PATH)
```

```python
df.head()
```

Setting the theme

```python
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 100
```

Checking the histogram to identify the count or diveristy across varied temperature

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Tg'], kde=True, color='royalblue', bins=30)
plt.axvline(df['Tg'].mean(), color='red', linestyle='--', label=f'Mean: {df['Tg'].mean():.2f}')
plt.title(f'Distribution of Glass Transition Temperature (Count)', fontsize=15)
plt.xlabel('$T_g$ (Celsius)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
ax=sns.kdeplot(df['Tg'], bw_adjust=0.5, fill=True, color="teal")
ax.set_title('Distribution of Polymer Glass Transition Temperatures (Density)', fontsize=14, pad=15)
```
